## Funci贸n para evaluar una funci贸n de membres铆a triangular izquierda

$$
\text{MF\_triangularIzquierda}(x; a, b) =
\begin{cases}
1, & \text{si } x \leq a. \\
\frac{b - x}{b - a}, & \text{si } a \leq x \leq b. \\
0, & \text{si } b \leq x.
\end{cases}
$$

- **x** = Punto de evaluaci贸n.
- **a** = V茅rtice izquierdo de la funci贸n triangular.
- **b** = V茅rtice derecho de la funci贸n triangular.
- **min_universo** = Valor m铆nimo del conjunto universo.
- **max_universo** = Valor m谩ximo del conjunto universo.rso.

```python
import matplotlib.pyplot as plt
import numpy as np

def MF_triangularIzquierda(X, a, b):
    if X <= a:
        return 1
    elif a < X < b:
        return -(1 / (b - a)) * (X - a) + 1
    else:
        return 0

# Funci贸n para solicitar los valores de a, b y el rango del conjunto universo:
def ltMF_main():
    while True:
        a = float(input("Ingrese el valor de a: "))
        b = float(input("Ingrese el valor de b: "))
        if a < b:
            break
        else:
            print("Error: El valor de a debe ser menor que b. Por favor, ingrese los valores nuevamente.")

    while True:
        min_universo = float(input("Ingrese el valor m铆nimo del conjunto universo: "))
        max_universo = float(input("Ingrese el valor m谩ximo del conjunto universo: "))
        if min_universo < max_universo:
            break
        else:
            print("Error: El valor m铆nimo del conjunto universo debe ser menor al valor m谩ximo. Por favor, ingrese los valores nuevamente.")

    plot_MF_triangularIzquierda(a, b, min_universo, max_universo)

# Funci贸n para graficar la MF:
def plot_MF_triangularIzquierda(a, b, min_universo, max_universo):
    X_values = np.linspace(min_universo, max_universo, 500)
    X_values = np.concatenate(([a, b], X_values))
    X_values = np.sort(X_values)

    MF_values = [MF_triangularIzquierda(X, a, b) for X in X_values]

    plt.figure(figsize=(10, 6))
    plt.plot(X_values, MF_values, label=f'Triangular Membership Function (a={a}, b={b})', color='blue')

    plt.xlabel('X')
    plt.ylabel('Membership Value')
    plt.title('Left Triangular Membership Function')
    plt.legend()
    plt.grid(True)
    plt.ylim(-0.1, 1.1)
    plt.xlim(min_universo, max_universo)
    plt.show()

ltMF_main()
```

## Funci贸n para evaluar una funci贸n de membres铆a triangular

## Funci贸n para evaluar una funci贸n de membres铆a trapezoidal

$$
\text{MF\_trapezoidal}(x; a, b, c, d) =
\begin{cases}
0, & \text{si } x \leq a. \\
\frac{x - a}{b - a}, & \text{si } a \leq x \leq b. \\
1, & \text{si } b \leq x \leq c. \\
\frac{d - x}{d - c}, & \text{si } c \leq x \leq d. \\
0, & \text{si } d \leq x.
\end{cases}
$$

- **x** = Punto de evaluaci贸n.
- **a** = V茅rtice inferior izquierdo de la funci贸n trapezoidal.
- **b** = V茅rtice superior izquierdo de la funci贸n trapezoidal.
- **c** = V茅rtice superior derecho de la funci贸n trapezoidal.
- **d** = V茅rtice inferior derecho de la funci贸n trapezoidal.
- **min_universo** = Valor m铆nimo del conjunto universo.
- **max_universo** = Valor m谩ximo del conjunto universo.

```python
import matplotlib.pyplot as plt
import numpy as np

def MF_trapezoidal(X, a, b, c, d):
    if X <= a:
        return 0
    elif a < X <= b:
        return (X - a) / (b - a)
    elif b < X <= c:
        return 1
    elif c < X < d:
        return (d - X) / (d - c)
    else:
        return 0

# Funci贸n para solicitar los valores de a, b, c, d y el rango del conjunto universo:
def tpMF_main():
    while True:
        a = float(input("Ingrese el valor de a: "))
        b = float(input("Ingrese el valor de b: "))
        c = float(input("Ingrese el valor de c: "))
        d = float(input("Ingrese el valor de d: "))
        if a < b < c < d:
            break
        else:
            print("Error: Los valores deben cumplir con la condici贸n a < b < c < d. Por favor, ingrese los valores nuevamente.")

    while True:
        min_universo = float(input("Ingrese el valor m铆nimo del conjunto universo: "))
        max_universo = float(input("Ingrese el valor m谩ximo del conjunto universo: "))
        if min_universo < max_universo:
            break
        else:
            print("Error: El valor m铆nimo del conjunto universo debe ser menor al valor m谩ximo. Por favor, ingrese los valores nuevamente.")

    plot_MF_trapezoidal(a, b, c, d, min_universo, max_universo)

# Funci贸n para graficar la MF:
def plot_MF_trapezoidal(a, b, c, d, min_universo, max_universo):
    X_values = np.linspace(min_universo, max_universo, 500)
    MF_values = [MF_trapezoidal(X, a, b, c, d) for X in X_values]

    plt.figure(figsize=(10, 6))
    plt.plot(X_values, MF_values, label=f'Trapezoidal Membership Function (a={a}, b={b}, c={c}, d={d})', color='blue')

    plt.xlabel('X')
    plt.ylabel('Membership Value')
    plt.title('Trapezoidal Membership Function')
    plt.legend()
    plt.grid(True)
    plt.ylim(-0.1, 1.1)
    plt.xlim(min_universo, max_universo)
    plt.show()

tpMF_main()
```

## Funci贸n para evaluar una funci贸n de membres铆a de Cauchy

## Funci贸n para evaluar una funci贸n de membres铆a izquierda-derecha (L-R)

$$
\text{LR}(x; c, \alpha, \beta) =
\begin{cases}
F_L\left(\frac{c - x}{\alpha}\right), & x \leq c. \\
F_R\left(\frac{x - c}{\beta}\right), & x \geq c.
\end{cases}
$$

$$
\begin{aligned}
F_L(x) &= \max(0, \sqrt{1 - x^2}). \\
F_R(x) &= e^{-|x|^3}.
\end{aligned}
$$

- **x** = Punto de evaluaci贸n.
- **$\alpha$** = Define la dispersi贸n de la parte izquierda de la funci贸n.
- **$\beta$** = Define la dispersi贸n de la parte derecha de la funci贸n.
- **c** = Valor del centro de la funci贸n.
- **min_universo** = Valor m铆nimo del conjunto universo.
- **max_universo** = Valor m谩ximo del conjunto universo.erso.

```python
import matplotlib.pyplot as plt
import numpy as np

def MF_leftRight(X, alpha, beta, c):
    if X >= c:
        x1 = (X - c) / beta
        return np.exp(-abs(x1**3))
    else:
        x2 = (c - X) / alpha
        return max(0, np.sqrt(1 - x2**2))

# Funci贸n para solicitar los valores de c, a, b y el rango del conjunto universo:
def lrMF_main():
    c = float(input("Ingrese el valor del centro (c): "))
    alpha = float(input("Ingrese el valor de alpha: "))
    beta = float(input("Ingrese el valor de beta: "))

    while True:
        min_universo = float(input("Ingrese el valor m铆nimo del conjunto universo: "))
        max_universo = float(input("Ingrese el valor m谩ximo del conjunto universo: "))
        if min_universo < max_universo:
            break
        else:
            print("Error: El valor m铆nimo del conjunto universo debe ser menor al valor m谩ximo. Por favor, ingrese los valores nuevamente.")

    plot_MF_leftRight(alpha, beta, c, min_universo, max_universo)

# Funci贸n para graficar la MF:
def plot_MF_leftRight(alpha, beta, c, min_universo, max_universo):
    X_values = np.linspace(min_universo, max_universo, 500)
    MF_values = [MF_leftRight(X, alpha, beta, c) for X in X_values]

    plt.figure(figsize=(10, 6))
    plt.plot(X_values, MF_values, label=f'Triangular Membership Function (alpha={alpha}, beta={beta}, c={c})', color='blue')

    plt.xlabel('X')
    plt.ylabel('Membership Value')
    plt.title('Left-Right Membership Function')
    plt.legend()
    plt.grid(True)
    plt.ylim(-0.1, 1.1)
    plt.xlim(min_universo, max_universo)
    plt.show()

lrMF_main()
```

## Funci贸n de membres铆a triangular

```python
import matplotlib.pyplot as plt
import numpy as np

# Funci贸n para evaluar una MF triangular en 2D:
def MF_triangular_2D(X, Y, a, b, c, d, e, f):
    Z = np.zeros_like(X)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            if X[i, j] <= a or Y[i, j] <= d:
                Z[i, j] = 0
            elif a < X[i, j] <= b and d < Y[i, j] <= e:
                Z[i, j] = min((X[i, j] - a) / (b - a), (Y[i, j] - d) / (e - d))
            elif b < X[i, j] < c and e < Y[i, j] < f:
                Z[i, j] = min(-(X[i, j] - c) / (c - b), -(Y[i, j] - f) / (f - e))
            else:
                Z[i, j] = 0
    return Z

# Funci贸n para solicitar los valores de los par谩metros y el rango del conjunto universo:
def tnMF_3D_main():
    while True:
        a = float(input("Ingrese el valor de a (para X): "))
        b = float(input("Ingrese el valor de b (para X): "))
        c = float(input("Ingrese el valor de c (para X): "))
        if a < b < c:
            break
        else:
            print("Error: El valor de a debe ser menor que b y el valor de b debe ser menor que c. Por favor, ingrese los valores nuevamente.")

    while True:
        d = float(input("Ingrese el valor de d (para Y): "))
        e = float(input("Ingrese el valor de e (para Y): "))
        f = float(input("Ingrese el valor de f (para Y): "))
        if d < e < f:
            break
        else:
            print("Error: El valor de d debe ser menor que e y el valor de e debe ser menor que f. Por favor, ingrese los valores nuevamente.")

    while True:
        min_universo_x = float(input("Ingrese el valor m铆nimo del conjunto universo para X: "))
        max_universo_x = float(input("Ingrese el valor m谩ximo del conjunto universo para X: "))
        min_universo_y = float(input("Ingrese el valor m铆nimo del conjunto universo para Y: "))
        max_universo_y = float(input("Ingrese el valor m谩ximo del conjunto universo para Y: "))
        if min_universo_x < max_universo_x and min_universo_y < max_universo_y:
            break
        else:
            print("Error: El valor m铆nimo del conjunto universo debe ser menor al valor m谩ximo para ambos ejes. Por favor, ingrese los valores nuevamente.")

    plot_MF_triangular_3D(a, b, c, d, e, f, min_universo_x, max_universo_x, min_universo_y, max_universo_y)

# Funci贸n para graficar la MF en 3D:
def plot_MF_triangular_3D(a, b, c, d, e, f, min_universo_x, max_universo_x, min_universo_y, max_universo_y):
    X = np.linspace(min_universo_x, max_universo_x, 500)
    Y = np.linspace(min_universo_y, max_universo_y, 500)
    X, Y = np.meshgrid(X, Y)
    Z = MF_triangular_2D(X, Y, a, b, c, d, e, f)

    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Membership Value')
    ax.set_title('Triangular Membership Function in 3D')
    plt.show()

# Llamar la funci贸n principal
tnMF_3D_main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Funci贸n para evaluar la MF Izquierda-Derecha en 2D:
def MF_leftRight_2D(X, Y, alpha, beta, c):
    Z = np.zeros_like(X)
    for i in range(X.shape[0]):
        for j in range(X.shape[1]):
            if X[i, j] >= c:
                x1 = (X[i, j] - c) / beta
                Z[i, j] = np.exp(-abs(x1**3))
            else:
                x2 = (c - X[i, j]) / alpha
                Z[i, j] = max(0, np.sqrt(1 - x2**2))
    return Z

# Funci贸n para solicitar los valores de los par谩metros y el rango del conjunto universo:
def leftRightMF_3D_main():
    alpha = float(input("Ingrese el valor del par谩metro alpha (define la dispersi贸n izquierda): "))
    beta = float(input("Ingrese el valor del par谩metro beta (define la dispersi贸n derecha): "))
    c = float(input("Ingrese el valor del centro c: "))

    while True:
        min_universo_x = float(input("Ingrese el valor m铆nimo del conjunto universo para X: "))
        max_universo_x = float(input("Ingrese el valor m谩ximo del conjunto universo para X: "))
        min_universo_y = float(input("Ingrese el valor m铆nimo del conjunto universo para Y: "))
        max_universo_y = float(input("Ingrese el valor m谩ximo del conjunto universo para Y: "))
        if min_universo_x < max_universo_x and min_universo_y < max_universo_y:
            break
        else:
            print("Error: El valor m铆nimo del conjunto universo debe ser menor al valor m谩ximo para ambos ejes. Por favor, ingrese los valores nuevamente.")

    plot_MF_leftRight_3D(alpha, beta, c, min_universo_x, max_universo_x, min_universo_y, max_universo_y)

# Funci贸n para graficar la MF Izquierda-Derecha en 3D:
def plot_MF_leftRight_3D(alpha, beta, c, min_universo_x, max_universo_x, min_universo_y, max_universo_y):
    X = np.linspace(min_universo_x, max_universo_x, 500)
    Y = np.linspace(min_universo_y, max_universo_y, 500)
    X, Y = np.meshgrid(X, Y)
    Z = MF_leftRight_2D(X, Y, alpha, beta, c)

    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='viridis')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Membership Value')
    ax.set_title('Left-Right Membership Function in 3D')
    plt.show()

# Llamar la funci贸n principal
leftRightMF_3D_main()
```
