# Conjuntos y sistemas difusos tipo II

Los conjuntos difusos "tradicionales" o tipo I (T1 FS) han existido desde 1965 y han sido utilizados con éxito en muchas aplicaciones; sin embargo, estos conjuntos tienen capacidades limitadas para manejar directamente las incertidumbres en los datos, donde manejar significa modelar y minimizar el efecto de estas incertidumbres. Que un conjunto difuso tipo I no pueda hacer esto suena paradójico, porque la palabra difuso connota incertidumbre. Esta paradoja se conoce desde hace mucho tiempo, pero es cuestionable quién fue el primero en referirse a "difuso" como paradójico.

Por supuesto, la incertidumbre se presenta de muchas maneras y es independiente del tipo de conjunto difuso o metodología utilizada para manejarla. Dos tipos importantes de incertidumbre son la lingüística y la aleatoria. La primera está asociada con las palabras y con el hecho de que las palabras pueden significar cosas diferentes para diferentes personas, mientras que la segunda está asociada con la impredecibilidad. La teoría de la probabilidad se usa para manejar la incertidumbre aleatoria, y los conjuntos difusos se utilizan para manejar la incertidumbre lingüística. A veces, los conjuntos difusos también pueden manejar ambos tipos de incertidumbre, ya que un sistema difuso puede tener mediciones ruidosas u operar bajo perturbaciones aleatorias.

Dentro de la teoría de la probabilidad, se parte de una función de densidad de probabilidad (pdf) que incorpora toda la información sobre las incertidumbres aleatorias; sin embargo, en la mayoría de las aplicaciones prácticas, es imposible conocer o determinar la función de densidad de probabilidad. El hecho de que una pdf esté completamente caracterizada por todos sus momentos lo demuestra. Para la mayoría de las pdf, se requiere un número infinito de momentos; desafortunadamente, en la práctica no es posible determinar un número infinito de momentos, por lo que en su lugar se calculan suficientes momentos para extraer la mayor cantidad de información posible de los datos. Como mínimo, se utilizan dos momentos: la media y la varianza. Usar solo momentos de primer orden no sería muy útil, porque la incertidumbre aleatoria requiere comprender la dispersión en torno a la media, y esta información la proporciona la varianza. Así, el enfoque probabilístico aceptado para la incertidumbre aleatoria se basa, en gran medida, en métodos que utilizan al menos los dos primeros momentos de una pdf. Este es el motivo, por ejemplo, por el cual los diseños basados en la minimización del error cuadrático medio son tan populares.

Al igual que la varianza proporciona una medida de dispersión sobre la media, un conjunto difuso también necesita alguna medida de dispersión para capturar más incertidumbre que la que proporciona solo una función de membresía simple (MF), que es todo lo que se obtiene cuando se usa un conjunto difuso tipo I. Un conjunto difuso tipo II proporciona esta medida de dispersión.

## Conjuntos difusos tipo II y cómo se diferencian de un conjunto difuso tipo I

Un conjunto difuso tipo I tiene un grado de pertenencia que es certero, mientras que un conjunto difuso tipo II tiene grados de pertenencia que son difusos, por lo que podría llamarse un "conjunto difuso-difuso". Este conjunto es útil en circunstancias donde es difícil determinar la función de membresía exacta de un conjunto difuso, como en el modelado por conjuntos difusos.

Como ejemplo, supongamos que la variable de interés es el contacto visual, denotado por $x$, donde $x \in [0,10]$, y esta es una intensidad que en 0 denota la ausencia de contacto visual y en 10 denota la máxima cantidad de contacto visual. Uno de los términos que podría caracterizar la cantidad de contacto visual percibido (por ejemplo, durante la inspección en un aeropuerto) es "algo de contacto visual". Supongamos que 50 hombres y mujeres son encuestados y se les pide ubicar los extremos de un intervalo para algo de contacto visual en la escala 0-10. Seguramente, no se obtendrán los mismos resultados para todos, ya que las palabras significan cosas diferentes para diferentes personas.

Un enfoque para el uso de los 50 conjuntos de dos puntos extremos es promediar los datos de los extremos y luego usar los valores promedio para construir un intervalo asociado con algo de contacto visual. Se puede construir una función de membresía triangular (aunque se pueden usar otras formas), $MF(x)$, donde la base y el ápice de la función de membresía están en el eje $x$ en los valores promedios de los puntos extremos y cuyo vértice está en el punto medio entre los dos extremos. Esta función de membresía triangular tipo I puede mostrarse en dos dimensiones; desafortunadamente, ha ignorado por completo las incertidumbres asociadas con los dos puntos extremos.

Un segundo enfoque es hacer uso de los valores promedio de los extremos y la desviación estándar de cada punto extremo para establecer un intervalo de incertidumbre alrededor de cada punto extremo promedio. Al hacer esto, podemos conocer la ubicación de los dos puntos extremos a lo largo del eje x como difuminados. Los triángulos pueden escalarse para que sus extremos base puedan estar en cualquier punto dentro de los intervalos a lo largo del eje x asociados con los valores promedio borrosos. Esto conduce a un continuo de funciones de membresía triangulares a lo largo del eje x. 

Para propósitos de esta discusión, supongamos que hay $N$ de tales triángulos. Entonces, en cada valor de $x$, puede haber hasta $N$ grados de función de membresía, $MF_1(x)$, $MF_2(x)$, ..., $MF_N(x)$. Cada uno de los posibles grados de función de membresía tiene un peso asignado, digamos $w_1(x)$, $w_2(x)$, ..., $w_N(x)$. Estos pesos pueden interpretarse como las posibilidades asociadas con el grado de cada triángulo en ese valor de $x$. En consecuencia, en cada $x$, la colección de grados es una función $\{(MF_i(x), w_i(x)), i = 1, ..., N\}$, llamada función de membresía secundaria. La función de membresía resultante tipo II es tridimensional.

Si toda la incertidumbre desaparece, entonces un conjunto difuso tipo II se reduce a un conjunto difuso tipo I, es decir, si las incertidumbres sobre los extremos izquierdo y derecho desaparecen, entonces solo sobrevive la función de membresía triangular. Esto es similar a lo que sucede en probabilidad, cuando la incertidumbre sobre una distribución desaparece y la función de densidad de probabilidad se reduce a una función determinista bien definida. En este caso, un conjunto difuso tipo I es simplemente un caso especial de un conjunto difuso tipo II, cuando no existe incertidumbre.

### Representaciones importantes de un conjunto difuso tipo II (T2 FS)

¿Existen representaciones importantes de un conjunto difuso tipo II y, si es así, por qué son importantes? Hay dos representaciones muy importantes para un conjunto difuso tipo II; están resumidas en la siguiente tabla:

| Nombre de la Representación       | Expresión Matemática                                      | Comentarios                                                                 |
|-----------------------------------|-----------------------------------------------------------|----------------------------------------------------------------------------|
| **Representación por cortes verticales** | $\tilde{A} = \bigcup\limits_{x \in X} \text{Cortes verticales} (x)$ | Muy útil para cálculos computacionales.                                   |
| **Representación por cortes ondulados** | $\tilde{A} = \bigcup\limits_{j} \text{FS T2 embebidos} (j)$ | Muy útil para derivaciones teóricas; también conocida como el *Teorema de Representación de Mendel-John*. |

La representación de corte vertical es la base para la mayoría de los cálculos, mientras que la representación de corte ondulado es la base de la mayoría de las derivaciones teóricas. Para un conjunto difuso tipo II, ambas representaciones también pueden interpretarse como teoremas de cobertura, ya que la unión de todos los cortes verticales y la unión de todos los conjuntos difusos tipo I embebidos cubren toda el área de incertidumbre (FOU, por sus siglas en inglés, *Field of Uncertainity*).

Aunque el Teorema de Representación de Mendel-John (RT) es extremadamente útil para desarrollos teóricos, no es muy útil para la computación porque el número de conjuntos embebidos en la unión puede ser astronómico de modo que, típicamente, el RT se usa para llegar a la estructura de un resultado teórico (por ejemplo, la unión de dos IT2 FS), mientras que los algoritmos computacionales prácticos se desarrollan para computar la estructura.

Para un IT2 FS, el RT establece que un IT2 FS es la unión de todos los T1 FS embebidos que cubren su FOU. La importancia de este resultado radica en que nos permite derivar todo sobre IT2 FS o sistemas utilizando matemáticas de T1 FS. Esto resulta en un enorme ahorro de tiempo.

## Sistemas de Lógica Difusa Tipo-2 (FLS)

*¿Cómo y por qué se utilizan los T2 FS en un sistema basado en reglas?*

Un FLS basado en reglas contiene cuatro componentes: reglas, difusificación, motor de inferencia y procesador de salida, los cuales están interconectados. Una vez establecidas las reglas, el FLS puede verse como una asignación de entradas a salidas, y esta asignación puede expresarse cuantitativamente como $y = f(x)$. Este tipo de FLS es ampliamente utilizado en muchas aplicaciones de ingeniería de lógica difusa (FL), como en controladores de FL y procesadores de señal, y también se conoce como controlador difuso o sistema difuso.

Las reglas son el núcleo de un FLS. Pueden ser proporcionadas por expertos o extraídas de datos numéricos. En cualquier caso, las reglas pueden expresarse como una colección de declaraciones SI-ENTONCES. Por ejemplo: SI la tasa promedio de entrada de datos de voz y video en tiempo real es *moderada*, y la tasa promedio de entrada de datos no en tiempo real es *alguna*, ENTONCES la confianza para aceptar la llamada telefónica es *alta*.

La parte SI de la regla se llama **antecedente**, y la parte ENTONCES de la regla se llama **consecuente**. Los FS están asociados con términos que aparecen en los antecedentes o consecuentes de las reglas y con las entradas y salidas del FLS. Las funciones de membresía (MFs) se utilizan para describir estos FS y pueden ser de tipo I o II. Estas últimas permiten cuantificar diferentes tipos de incertidumbre en un FLS.

Cuatro maneras en que la incertidumbre puede ocurrir en un FLS son las siguientes:
1. Las palabras utilizadas en los antecedentes y consecuentes de las reglas pueden tener diferentes significados para distintas personas.
2. Los consecuentes obtenidos al encuestar a un grupo de expertos a menudo diferirán para la misma regla, ya que los expertos no necesariamente estarán de acuerdo.
3. Solo hay datos de entrenamiento ruidosos disponibles para ajustar (optimizar) los parámetros de un IT2 FLS.
4. Las mediciones ruidosas activan el FLS.

Un FLS que se describe completamente en términos de T1 FS se denomina **T1 FLS**, mientras que un FLS descrito utilizando al menos un T2 FS se llama **T2 FLS**. Los T1 FLS son incapaces de manejar directamente estas incertidumbres porque usan T1 FS que son ciertos; por otro lado, los T2 FLS son muy útiles en circunstancias donde es difícil determinar una MF exacta para un término, por lo que pueden usarse para manejar estas incertidumbres.

### Cómo se pueden usar las matemáticas de T1 FS para derivar las salidas activadas por reglas de IT2 FS

Para ver el panorama general, centrémonos en la regla única *"Si $x$ es $\tilde{F}^1$ ENTONCES $y$ es $\tilde{G}^1$"*. Esta regla tiene un antecedente y un consecuente, y se activa por un número nítido (es decir, difusificación singleton). La clave para utilizar matemáticas de T1 FS para derivar la salida activada de IT2 FS es construir un gráfico. 

Obsérvese que el antecedente se descompone en sus $n_F$ FS embebidos de T1 y el consecuente se descompone en sus $n_G$ FS embebidos de T1. Cada uno de los $n_F \times n_G$ caminos actúa como una inferencia T1. Cuando se toma la unión de todos los conjuntos activados de T1, el resultado es el conjunto activado de T2. Este último tiene límites inferior y superior, porque cada uno de los conjuntos activados de T1 está acotado. La manera exacta de obtener una fórmula para $B(y)$ solo requiere calcular estos límites inferior y superior, los cuales solo involucran las funciones de membresía inferior y superior de los FOU del antecedente y del consecuente.

### Comparación entre la desdifusificación COS y TR

**Desfusificación del Centro de Conjuntos (COS), $y_{cos}(x)$**

1. Calcular el centroide del FS consecuente de cada regla T1 FS. Denótese como $c^l$ (donde $l = 1, ..., M$).
2. Calcular el nivel de activación de cada regla disparada. Denótese como $f^l$ (donde $l = 1, ..., M$).
3. Calcular $y_{cos}(x)$ como:

$$
y_{cos}(x) = \frac{\sum_{l=1}^{M} c^l f^l}{\sum_{l=1}^{M} f^l}
$$

**TR del Centro de Conjuntos (COS), $Y_{cos}(x)$**

1. Calcular el centroide del FS consecuente de cada regla IT2 FS, utilizando los algoritmos KM (ver Cuadro 6). Denótese como $[y_l^l, y_r^l]$ (donde $l = 1, ..., M$).
2. Calcular el intervalo de activación para cada regla disparada. Denótese como $[f_l^l, f_r^l]$ (donde $l = 1, ..., M$).
3. Computar $Y_{cos}(x) = [y_l(x), y_r(x)]$, donde $y_l(x)$ es la solución del siguiente problema de minimización:

$$
y_l(x) = \min_{y \in \{y_l^l, y_r^l\}} \frac{\sum_{l=1}^{M} y f^l_l}{\sum_{l=1}^{M} f^l_l}
$$

que se resuelve utilizando un Algoritmo KM (Cuadro 6), y \( y_r(x) \) es la solución del siguiente problema de maximización:

$$
y_r(x) = \max_{y \in \{y_l^l, y_r^l\}} \frac{\sum_{l=1}^{M} y f^l_r}{\sum_{l=1}^{M} f^l_r}
$$

que se resuelve utilizando el otro Algoritmo KM.

### Centroide de un IT2 FS y su cálculo

Usando el RT, se calculan los centroides de todos los T1 FS embebidos, ejemplos de los cuales se muestran como funciones coloreadas. Como el conjunto de centroides es finito, este cálculo lleva a un conjunto de centroides llamado el centroide de $\tilde{B}$, $C(\tilde{B}^{\%})$. 

$C(\tilde{B}^{\%})$ tiene un valor $c_l$ más pequeño y un valor $c_r$ más grande, es decir, $C(\tilde{B}^{\%}) = [c_l, c_r]$. Por lo tanto, para calcular $C(\tilde{B}^{\%})$, solo es necesario computar $c_l$ y $c_r$. No es posible hacerlo de forma cerrada. En su lugar, se pueden calcular $c_l$ y $c_r$ usando dos algoritmos iterativos, llamados algoritmos de Karnik-Mendel (KM). Nótese que $c_l = \min$ (centroide de todos los T1 FS embebidos en $FOU(\tilde{B}^{\%})$). El análisis muestra que:

$$
c_l = c_l(L) = \frac{\sum_{i=1}^{L} y_i UMF(\tilde{B}|y_i) + \sum_{i=L+1}^{N} y_i LMF(\tilde{B}|y_i)}{\sum_{i=1}^{L} UMF(\tilde{B}|y_i) + \sum_{i=L+1}^{N} LMF(\tilde{B}|y_i)}
$$

Uno de los algoritmos KM calcula el punto de cambio $L$. También se debe notar que $c_r = \max$ (centroide de todos los T1 FS embebidos en $FOU(\tilde{B}^{\%})$). El análisis muestra que:

$$
c_r = c_r(R) = \frac{\sum_{i=1}^{R} y_i LMF(\tilde{B}|y_i) + \sum_{i=R+1}^{N} y_i UMF(\tilde{B}|y_i)}{\sum_{i=1}^{R} LMF(\tilde{B}|y_i) + \sum_{i=R+1}^{N} UMF(\tilde{B}|y_i)}
$$

El otro algoritmo KM calcula el punto de cambio $R$. Un ejemplo del algoritmo KM se muestra a continuación.

```python
import numpy as np
import matplotlib.pyplot as plt

# Función de Karnik-Mendel para la desdifusificación
def Karnik_Mendel(a, b, c, d):
    n = len(a)

    # Ordenar "a"
    sorted_indices = np.argsort(a)
    a_sorted = np.array(a)[sorted_indices]
    c_sorted = np.array(c)[sorted_indices]
    d_sorted = np.array(d)[sorted_indices]

    # Algoritmo KM para y_L
    w_l = (c_sorted + d_sorted) / 2
    y_L = np.sum(a_sorted * w_l) / np.sum(w_l)

    return y_L

# Defición del rango de valores
n = 21
x = np.linspace(-5, 5, n)
y = np.linspace(-5, 5, n)

# Definición de parámetros
aA1_l, cA1_l = 2.2, -0.4
aA2_l, cA2_l = 2.2, 0.6
aA1_r, cA1_r = 1.8, 0.2
aA2_r, cA2_r = 1.8, 0.2

aB1_l, cB1_l = 2.2, -0.4
aB2_l, cB2_l = 2.2, 0.6
aB1_r, cB1_r = 1.8, 0.2
aB2_r, cB2_r = 1.8, 0.2

# Inicialización de matrices
A1_l = 1 / (1 + np.exp(aA1_l * (x - cA1_l)))
A2_l = 1 / (1 + np.exp(-aA2_l * (x - cA2_l)))
A1_r = 1 / (1 + np.exp(aA1_r * (x - cA1_r)))
A2_r = 1 / (1 + np.exp(-aA2_r * (x - cA2_r)))

B1_l = 1 / (1 + np.exp(aB1_l * (y - cB1_l)))
B2_l = 1 / (1 + np.exp(-aB2_l * (y - cB2_l)))
B1_r = 1 / (1 + np.exp(aB1_r * (y - cB1_r)))
B2_r = 1 / (1 + np.exp(-aB2_r * (y - cB2_r)))

F = np.zeros((n, n))

# Cálculo de F(r, p)
for r in range(n):
    for p in range(n):
        w1_l = A1_l[r] * B1_l[p]
        w2_l = A1_l[r] * B2_l[p]
        w3_l = A2_l[r] * B1_l[p]
        w4_l = A2_l[r] * B2_l[p]
        w_l = w1_l + w2_l + w3_l + w4_l + 0.0001

        w1_r = A1_r[r] * B1_r[p]
        w2_r = A1_r[r] * B2_r[p]
        w3_r = A2_r[r] * B1_r[p]
        w4_r = A2_r[r] * B2_r[p]
        w_r = w1_r + w2_r + w3_r + w4_r + 0.0001

        w_1_l = w1_l / w_l
        w_2_l = w2_l / w_l
        w_3_l = w3_l / w_l
        w_4_l = w4_l / w_l

        w_1_r = w1_r / w_r
        w_2_r = w2_r / w_r
        w_3_r = w3_r / w_r
        w_4_r = w4_r / w_r

        f1_low = -0.5 * x[r] + 0.5 * y[p] + 0.5
        f2_low = -0.5 * y[p] + 2.5
        f3_low = -0.5 * x[r] + 2.5
        f4_low = 0.5 * x[r] + 0.5 * y[p] + 1.5

        f1_up = -1.5 * x[r] + 1.5 * y[p] + 1
        f2_up = -1.5 * y[p] + 3.5
        f3_up = -1.5 * x[r] + 3.5
        f4_up = 1.5 * x[r] + 1.5 * y[p] + 2.5

        a = [f1_low, f2_low, f3_low, f4_low]
        b = [f1_up, f2_up, f3_up, f4_up]
        f_low = [w_1_l, w_2_l, w_3_l, w_4_l]
        f_up = [w_1_r, w_2_r, w_3_r, w_4_r]

        F[r, p] = Karnik_Mendel(a, b, f_low, f_up)

# Graficar funciones de membresía
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x, A1_l, 'r', label="A1_l")
plt.plot(x, A2_l, 'r', linestyle='dashed', label="A2_l")
plt.plot(x, A1_r, 'b', label="A1_r")
plt.plot(x, A2_r, 'b', linestyle='dashed', label="A2_r")
plt.legend()
plt.title("Funciones de Membresía A")
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(y, B1_l, 'r', label="B1_l")
plt.plot(y, B2_l, 'r', linestyle='dashed', label="B2_l")
plt.plot(y, B1_r, 'b', label="B1_r")
plt.plot(y, B2_r, 'b', linestyle='dashed', label="B2_r")
plt.legend()
plt.title("Funciones de Membresía B")
plt.grid()

plt.show()

# Gráfico de superficie
X, Y = np.meshgrid(x, y)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, F, cmap='viridis')
ax.set_title("Superficie de F")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("F")
plt.show()
```

Alternativamente, es posible utilizar el algoritmo de Wu-Mendel (WM), un ejemplo se muestra a continuación.

```python
import numpy as np
import matplotlib.pyplot as plt

# Función de Wu-Mendel
def Wu_Mendel_Uncertainty_Bounds(a, b, f_low, f_up):
    M = len(a)
    
    y_l_0 = np.sum(f_low * a) / np.sum(f_low)
    y_r_M = np.sum(f_low * b) / np.sum(f_low)
    y_l_M = np.sum(f_up * a) / np.sum(f_up)
    y_r_0 = np.sum(f_up * b) / np.sum(f_up)

    y_l_up = min(y_l_0, y_l_M)
    y_r_low = max(y_r_0, y_r_M)

    A = np.sum(f_up - f_low)
    B = np.sum(f_up) * np.sum(f_low)
    C = np.sum(f_low * (a - a[0]))
    D = np.sum(f_up * (a[-1] - a))
    E = np.sum(f_up * (b - b[0]))
    F = np.sum(f_low * (b[-1] - b))

    y_l_low = y_l_up - ((A / B) * ((C * D) / (C + D)))
    y_r_up = y_r_low + ((A / B) * ((E * F) / (E + F)))

    y_l = (y_l_low + y_l_up) / 2
    y_r = (y_r_low + y_r_up) / 2
    
    return (y_l + y_r) / 2

# Definición del rango de valores
n = 21
x = np.linspace(-5, 5, n)
y = np.linspace(-5, 5, n)

# Definición de parámetros
aA1_l, cA1_l = 2.2, -0.4
aA2_l, cA2_l = 2.2, 0.6
aA1_r, cA1_r = 1.8, 0.2
aA2_r, cA2_r = 1.8, 0.2

aB1_l, cB1_l = 2.2, -0.4
aB2_l, cB2_l = 2.2, 0.6
aB1_r, cB1_r = 1.8, 0.2
aB2_r, cB2_r = 1.8, 0.2

# Inicialización de matrices
A1_l = 1 / (1 + np.exp(aA1_l * (x - cA1_l)))
A2_l = 1 / (1 + np.exp(-aA2_l * (x - cA2_l)))
A1_r = 1 / (1 + np.exp(aA1_r * (x - cA1_r)))
A2_r = 1 / (1 + np.exp(-aA2_r * (x - cA2_r)))

B1_l = 1 / (1 + np.exp(aB1_l * (y - cB1_l)))
B2_l = 1 / (1 + np.exp(-aB2_l * (y - cB2_l)))
B1_r = 1 / (1 + np.exp(aB1_r * (y - cB1_r)))
B2_r = 1 / (1 + np.exp(-aB2_r * (y - cB2_r)))

F = np.zeros((n, n))

# Cálculo de F
for r in range(n):
    for p in range(n):
        w1_l, w2_l, w3_l, w4_l = A1_l[r] * B1_l[p], A1_l[r] * B2_l[p], A2_l[r] * B1_l[p], A2_l[r] * B2_l[p]
        w_l = w1_l + w2_l + w3_l + w4_l + 1e-4
        
        w1_r, w2_r, w3_r, w4_r = A1_r[r] * B1_r[p], A1_r[r] * B2_r[p], A2_r[r] * B1_r[p], A2_r[r] * B2_r[p]
        w_r = w1_r + w2_r + w3_r + w4_r + 1e-4

        w_low = np.array([w1_l / w_l, w2_l / w_l, w3_l / w_l, w4_l / w_l])
        w_up = np.array([w1_r / w_r, w2_r / w_r, w3_r / w_r, w4_r / w_r])
        
        f_low = np.array([
            -0.5 * x[r] + 0.5 * y[p] + 0.5,
            -0.5 * y[p] + 2.5,
            -0.5 * x[r] + 2.5,
            0.5 * x[r] + 0.5 * y[p] + 1.5
        ])
        
        f_up = np.array([
            -1.5 * x[r] + 1.5 * y[p] + 1,
            -1.5 * y[p] + 3.5,
            -1.5 * x[r] + 3.5,
            1.5 * x[r] + 1.5 * y[p] + 2.5
        ])

        F[r, p] = Wu_Mendel_Uncertainty_Bounds(f_low, f_up, w_low, w_up)

# Graficar las funciones de membresía
plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(x, A1_l, 'r', label="A1_l")
plt.plot(x, A2_l, 'r', linestyle='dashed', label="A2_l")
plt.plot(x, A1_r, 'b', label="A1_r")
plt.plot(x, A2_r, 'b', linestyle='dashed', label="A2_r")
plt.legend()
plt.title("Funciones de Membresía A")
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(y, B1_l, 'r', label="B1_l")
plt.plot(y, B2_l, 'r', linestyle='dashed', label="B2_l")
plt.plot(y, B1_r, 'b', label="B1_r")
plt.plot(y, B2_r, 'b', linestyle='dashed', label="B2_r")
plt.legend()
plt.title("Funciones de Membresía B")
plt.grid()

plt.show()

# Graficar la superficie
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(x, y)
ax.plot_surface(X, Y, F.T, cmap="viridis")
ax.set_title("Wu-Mendel Fuzzy System")
plt.show()
```

## Casi todas las aplicaciones utilizan IT2 FSs porque todos los cálculos son fáciles de realizar

Es necesario activar reglas que estén en términos de variables lingüísticas, las cuales tienen FSs asociados. Las entradas del FLS antes de la difusificación pueden ser ciertas (por ejemplo, mediciones perfectas) o inciertas (por ejemplo, mediciones ruidosas). La función de membresía para un T2 FS nos permite manejar cualquiera de estos tipos de medición. 

Los procedimientos inferenciales más comúnmente utilizados para un FLS usan modelos de implicación de *mínimo* y *producto*, donde los T2 FSs resultantes se llaman *T2 FLSs de Mamdani*. También están disponibles los *T2 FLSs TSK*.

### Límites de Incertidumbre y Cómputos Relacionados

**1. Cálculo de los Centroides de $M$ Consecuentes IT2 FSs**

Los puntos extremos de los centroides de los $M$ consecuentes IT2 FSs, $y_i$ y $y_i' (i = 1, ..., M)$, se calculan utilizando los algoritmos KM. Estos cálculos pueden realizarse después de que el diseño del IT2 FLS se haya completado y solo deben realizarse una vez.

**2. Cálculo de los cuatro centroides tipo I del límite**

$$
y_r^{(0)}(x) = \frac{\sum_{i=1}^{M} f_i y_i}{\sum_{i=1}^{M} f_i}
$$

$$
y_r^{(M)}(x) = \frac{\sum_{i=1}^{M} f_i' y_i}{\sum_{i=1}^{M} f_i'}
$$

$$
y_l^{(0)}(x) = \frac{\sum_{i=1}^{M} f_i y_i'}{\sum_{i=1}^{M} f_i}
$$

$$
y_l^{(M)}(x) = \frac{\sum_{i=1}^{M} f_i' y_i'}{\sum_{i=1}^{M} f_i'}
$$

**3. Cálculo de los Cuatro Límites de Incertidumbre**

$$
y_l(x) \leq y_r(x) \leq \bar{y}_r(x)
$$

$$
\underline{y}_l(x) \leq y_l(x)
$$

$$
\underline{y}_l(x) = \min \left[ y_l^{(0)}(x), y_l^{(M)}(x) \right]
$$

$$
\bar{y}_r(x) = \max \left[ y_r^{(0)}(x), y_r^{(M)}(x) \right]
$$

$$
y_l(x) = \underline{y}_l(x) + \left( \frac{\sum_{i=1}^{M} (f_i' - f_i)}{\sum_{i=1}^{M} f_i + \sum_{i=1}^{M} (f_i' - f_i)} \right) \times \left( \frac{\sum_{i=1}^{M} f_i (y_i' - y_i)}{\sum_{i=1}^{M} f_i} + \frac{\sum_{i=1}^{M} f_i' (y_i^M - y_i')}{\sum_{i=1}^{M} f_i'} \right)
$$

Nota: El par $\underline{y}_l(x), \bar{y}_r(x)$ se llama límite de incertidumbre interno, mientras que el par $y_l(x), y_r(x)$ se llama límite de incertidumbre externo.

**4. Cálculo del Conjunto TR Aproximado**

$$
y_{TR}(x) = [\underline{y}_l(x), \bar{y}_r(x)] = [y_l(x), y_r(x)]
$$

**5. Cálculo del Salida Difusa Aproximada**

$$
y(x) \approx p(x) = \frac{1}{2} [y_l(x) + y_r(x)]
$$

## Existen dos representaciones muy importantes para un T2 FS: representaciones de corte vertical y de corte ondulado

En muchas aplicaciones de un FLS, se deben obtener números precisos en su salida. Esto se logra mediante el procesador de salida, y se conoce como *desdifusificación*. En una aplicación de control, por ejemplo, dicho número corresponde a una acción de control; en una aplicación de procesamiento de señales, dicho número podría corresponder a un pronóstico financiero o la ubicación de un objetivo. El procesador de salida de un FLS T1 es solo un desdifusificador; sin embargo, el procesador de salida de un FLS T2 contiene dos componentes: el primero mapea un T2 FS a un T1 FS y se llama *reductor de tipo* (que realiza la reducción de tipo, TR), y el segundo realiza la desdifusificación sobre el conjunto reducido.

Hasta la fecha, los cálculos prácticos para un FLS T2 solo son posibles cuando los T2 FS son IT2 FS, es decir, para un IT2 FLS.

### Cálculos en un IT2 FLS

*¿Cuáles son los cálculos detallados para un IT2 FLS y son fáciles de entender?* 

Como se mencionó anteriormente, el TR se puede utilizar para derivar las fórmulas de entrada-salida para un IT2 FLS. Cuanto mayores (menores) sean las incertidumbres, mayor (menor) será el intervalo de activación y mayor (menor) será el FOU de la regla activada.

La salida del bloque de inferencia es procesada a continuación por el Procesador de Salida, que consta de dos etapas: Reducción de Tipo (TR) y Desdifusificación. Todos los métodos de TR son extensiones T2 de los métodos de defuzzificación T1, cada uno de los cuales se basa en algún tipo de cálculo de centroide; por ejemplo, en un FLS T1, los conjuntos de salida activados primero se pueden combinar mediante una operación de unión, tras lo cual se puede calcular el centroide del T1 FS resultante. Esto se llama *desdifusificación de centroide*. Alternativamente, dado que la operación de unión es computacionalmente costosa, cada nivel de activación se puede combinar con el centroide de su conjunto consecuente mediante un cálculo de centroide diferente, llamado *desdifusificación del centro de conjuntos*.

Los análogos T2 de estos dos tipos de defuzzificación se denominan *centroide TR* y *centro de conjuntos TR*. El resultado del TR para los IT2 FSs es un conjunto de intervalo $[y_l(x), y_r(x)]$.

Independientemente del tipo de TR, todos requieren calcular el *centroide de un IT2 FS*. Los dos algoritmos iterativos para hacerlo se conocen como los *algoritmos Karnik-Mendel (KM)*, y tienen las siguientes propiedades:

1. Son muy simples.
2. Convergen a las soluciones exactas de manera monótona y exponencialmente rápida.
3. Pueden ejecutarse en paralelo, ya que son independientes.

El marco general de cálculos se resume en que los intervalos de activación se calculan para todos los conjuntos de reglas activadas, y dependen del conjunto de entrada $x$. Para el *centro de conjuntos TR*, se realizan cálculos fuera de línea de los centroides para cada uno de los $M$ conjuntos consecuentes IT2 FS mediante los algoritmos KM, y se almacenan en memoria. Los TR del tipo *centro de conjuntos* combinan los intervalos de activación y los conjuntos consecuentes precomputados, y utilizan los algoritmos KM para realizar los cálculos estáticos.

### Un IT2 FLS para Cálculos en Tiempo Real

*¿Es posible tener un IT2 FLS sin TR?*

El TR es un cuello de botella para las aplicaciones en tiempo real de un IT2 FLS porque utiliza los algoritmos iterativos KM para sus cálculos. Aunque estos algoritmos son muy rápidos, existe un retraso asociado con cualquier algoritmo iterativo. La buena noticia es que el TR puede ser reemplazado utilizando los *límites mínimos de incertidumbre* para $y_l(x)$ y $y_r(x)$. Estos límites también son conocidos como los *límites de incertidumbre Wu-Mendel*. Se calculan cuatro límites: los límites inferiores y superiores para ambos $y_l(x)$ y $y_r(x)$. Las fórmulas para estos límites solo dependen de los niveles de activación inferior y superior de cada regla y del centroide del conjunto consecuente de cada regla.

La parte inicial de los cálculos en un IT2 FLS que usa límites de incertidumbre es idéntica a la parte inicial cuando se usa TR. Después de que se calculan los límites de incertidumbre, los valores reales de $y_l(x)$ y $y_r(x)$ (que se habrían calculado utilizando TR) se aproximan promediando sus respectivos límites de incertidumbre, dando como resultado $\hat{y_l}(x)$ y $\hat{y_r}(x)$. Luego, la desdifusificación se logra promediando estas dos aproximaciones, resultando en $\hat{y}(x)$, que es una aproximación de $y(x)$. Sorprendentemente, se pierde muy poca precisión cuando se utilizan los límites de incertidumbre.

Resumiendo, los IT2 FLS son bastante simples de entender y pueden implementarse de dos maneras: una que usa TR y otra que utiliza límites de incertidumbre.
